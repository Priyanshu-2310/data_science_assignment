```python
"""Sort Python Dictionaries by key"""
p = {"apple": 1, "orange": 2, "mango": 2, "sugarcane": 4}
a = list(p.keys())
a.sort()
print(a)
sorted_dict = {i: p[i] for i in a}
print(sorted_dict)
```

```python
"""User input dictionary"""
n = int(input("Kitne items chhiye = "))
dic = {}
for x in range(n):
    k = input("enter key = ")
    value = input("enter val = ")
    dic.update({k: value})

print(dic)
```

```python
"""take the input from the user and reverse the dictionary"""

n = int(input("Kitne items chhiye = "))
dic = {}
for x in range(n):
    k = input("enter key = ")
    value = input("enter val = ")
    dic.update({k: value})

list1 = list(dic.keys())
list1.reverse()

newdict = {i : dic[i] for i in list1}
print(newdict)

```

4) Write a Python program to get the maximum and minimum values of a dictionary.

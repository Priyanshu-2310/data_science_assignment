1. Write a program to print n natural number in descending order using a while loop.

```python
num = int(input("Enter the number = "))
while(num >=1):
    print(num)
    num-=1
```

2. Write a program that appends the square of each number to a new list.
Given x = [2, 3, 4, 5, 6, 7, 8]
Expected output
Result: [4, 9, 16, 25, 36, 49, 64]


```python
list1 = [2, 3, 4, 5, 6, 7, 8]
ans = []

for x in list1:
    ans.append(x**2)

print(ans)
```

Write a program that appends the type of elements from a list.
Given x = [23, 'Python', 23.98]
Expected output
Result: [<class 'int'>, <class 'str'>, <class 'float'>]

```python
list1 = [23, 'python', 23.98]
typelist = []

for x in list1:
    typelist.append(type(x))

print(typelist)
```

Question 4 - Write a program to filter even and odd numbers from a list.
Given x = [10, 23, 24, 35, 65, 78, 90]
Expected output
Even numbers: [10, 24, 78, 90] Odd numbers: [23, 35, 65]

```python
list1 = [10, 23, 24, 35, 65, 78, 90]
eve = []
odd = []

for x in list1:
    if(x%2 == 0):
        eve.append(x)
    else:
        odd.append(x)

print(eve, odd)
```

Question 5 : Python program to convert the month name to a number of days

```python
month = input("Enter month name: ").lower()

if month in ("january", "march", "may", "july", "august", "october", "december"):
    print("31 days")
elif month in ("april", "june", "september", "november"):
    print("30 days")
elif month == "february":
    print("28 or 29 days")
else:
    print("Invalid month name")
```

Question 6 : Python program that accepts a word from the user and reverses it.

```python
word = input("Enter the word")
realstr = []

for x in word:
    realstr.append(x)

le = len(realstr)
reverse = " "
while(le>=1):
    reverse = reverse + realstr[le - 1]
    le -= 1
print(reverse)
```

Question 7 : Python program to count the total number of digits in a number.

```python
num = int(input("Enter your number = "))
count = 0
while(num != 0 ):
    num = int(num / 10)
    count += 1

print(count)
```

Question 8 : Find the factorial of a given number.


```python
num = int(input("Enter your number = "))
fac = 1
while(num >= 1):
    fac = fac * num
    num -= 1

print(fac)
```

Question 1 : Write a Python function to check whether a number falls within a given range 10 to 900.

```python
def checkNumber(n):
    if(n>=10 and n<=900):
        print("With in the range")
    else:
        print("Not in range")
```

```python
checkNumber(5)
```

Question 2 :Write a Python function to multiply all the numbers in a list.


```python
def listmul(listname, mul):
    newlist = []
    for x in listname:
        newlist.append(x*mul)

    return newlist
```

```python
listmul([2, 3, 4, 5], 5)
```

Question 3 :Write a Python program to print the even numbers from a given list.
Sample List: [1, 2, 3, 4, 5, 6, 7, 8, 9]
Expected Result: [2, 4, 6, 8]


```python
def evenfun(list1):
    newlist = []
    for x in list1:
        if(x % 2 == 0):
            newlist.append(x)

    return(newlist)
```

```python
evenfun( [1, 2, 3, 4, 5, 6, 7, 8, 9])
```

Question 4 :Write a Python function that takes a list and returns a new list with distinct elements from the first list.Sample List: [1, 2, 3, 3, 3, 3, 4, 5]
Unique List: [1, 2, 3, 4, 5]


```python
def uniqueitem(simpleList):
    uniquelist = []
    for x in simpleList:
       if x not in uniquelist:
            uniquelist.append(x)

    return uniquelist
         
                
```

```python
uniqueitem([1, 2, 3, 3, 3, 3, 4, 5])
```

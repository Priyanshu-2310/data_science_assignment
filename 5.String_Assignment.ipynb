1) Python program to print even length words in a string
Input:
s = "This is a python language"
Output:
This is python language

```python
str1 = "This is a python language"
list1 = str1.split()
listeven = []
for x in list1:
    if(len(x)%2 == 0):
        listeven.append(x)


finalstr = " ".join(listeven)
print(finalstr)
```

3) Python program to find Frequency of numbers in String
Output:
The original string is : pythonfor data analyst is No. 1 4 python
Count of numerics in string : 2

```python
str3 = "pythonfor data analyst is No. 1 4 python"
count = 0

for x in str3:
    if x.isdigit():
        count += 1

print(count)
```

5) Calculate the sum and average of the digits present in a string
Input:
input_str = "Datapython29@!8496"

Output:
Sum is: 38 Average is 6.333333333333333

```python
str4 = "Datapython29@!8496"
count = 0
sum1 = 0
for x in range(len(str4)):
    if(str4[x].isdigit()):
        sum1 = sum1 + int(str4[x])
        count+=1

avg = sum1/count
print(avg)
```

6) Count all letters, digits, and special symbols from a given string
Input: "A@un2st&#5ve"
Output:
Chars = 8 Digits = 2 Symbol = 3

```python
str1 = "A@un2st&#5ve"
chars = 0
digits = 0
symbol = 0

for x in range(len(str1)):
    if(str1[x].isalpha()):
       chars += 1
    elif(str1[x].isdigit()):
        digits += 1
    else:
        symbol += 1

print(chars, digits, symbol)
```

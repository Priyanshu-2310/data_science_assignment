1) Python program to interchange first and last elements in a list
Input: [12, 35, 9, 56, 24]
Output: [24, 35, 9, 56, 12]

```python
list1  = [12, 35, 9, 56, 24]

temp = list1[0]
list1[0] = list1[len(list1)-1]
list1[len(list1)-1] = temp

print(list1)
```

5) Program to print duplicates from a list of integers
Input: [1, 2, 1, 2, 3, 4, 5, 1, 1, 2, 5, 6, 7, 8, 9, 9]
Output: [1, 2, 5, 9]

```python
nums = [1, 2, 1, 2, 3, 4, 5, 1, 1, 2, 5, 6, 7, 8, 9, 9]

duplicates = []

for num in nums:
    if nums.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print(duplicates)
```

2) Write a Python program to find repeated items in a tuple.
Input (2, 4, 5, 6, 2, 3, 4, 4, 7, 4)
Output 4

```python
tuple2 = (2, 4, 5, 6, 2, 3, 4, 4, 7, 4)

max_count = 0
max_item = None

for x in tuple2:
    if tuple2.count(x) > max_count:
        max_count = tuple2.count(x)
        max_item = x

print("Maximum repeated item =", max_item)
print("It appears", max_count, "times.")
```
